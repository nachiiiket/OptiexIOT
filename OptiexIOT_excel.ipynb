```python
import numpy as np

# Keep DB order (timestamps decreasing), do NOT sort ascending
df = df.reset_index(drop=True)

df["Anomaly Flag"] = ""
df["Imputed Value"] = np.nan

n = len(df)
i = 1
while i < n - 1:
    prev = df.loc[i - 1, "Value"]
    curr = df.loc[i, "Value"]
    nxt  = df.loc[i + 1, "Value"]

    # ----------- RESET -----------
    # Drop (curr < prev) followed immediately by a value larger than prev -> Reset at curr
    if curr < prev and nxt > prev:
        if df.loc[i, "Anomaly Flag"] == "":
            df.loc[i, "Anomaly Flag"] = "Yes (Reset)"
        i += 1
        continue

    # ----------- FAULTY (consecutive zeros) -----------
    if curr == 0:
        start = i
        while i < n and df.loc[i, "Value"] == 0:
            i += 1
        end = i - 1

        # next non-zero exists?
        if i < n:
            after = df.loc[i, "Value"]
            # if after < prev => zeros are within the decreasing trend -> faulty
            if after < prev:
                imputed_val = (prev + after) / 2
                for j in range(start, end + 1):
                    df.loc[j, "Anomaly Flag"] = "Yes (Faulty)"
                    df.loc[j, "Imputed Value"] = imputed_val
            else:
                # after >= prev means this zero-block is more likely a Reset pattern
                # mark the first zero position as Reset (optional) â€” here we prefer Reset
                df.loc[start, "Anomaly Flag"] = "Yes (Reset)"
        else:
            # zeros reach end of file -> impute with prev
            imputed_val = prev
            for j in range(start, end + 1):
                df.loc[j, "Anomaly Flag"] = "Yes (Faulty)"
                df.loc[j, "Imputed Value"] = imputed_val
        continue

    # ----------- NEGATIVE DIFFERENCE -----------
    # Normal = decreasing. If next > curr and it's not a Reset (checked above), mark next as negative
    if nxt > curr:
        if df.loc[i + 1, "Anomaly Flag"] == "":
            df.loc[i + 1, "Anomaly Flag"] = "Yes (Negative Difference)"
        i += 1
        continue

    # otherwise move on
    i += 1

# final check: last row (no nxt available) could be a negative-diff vs previous
if n >= 2:
    if df.loc[n - 1, "Anomaly Flag"] == "" and df.loc[n - 1, "Value"] > df.loc[n - 2, "Value"]:
        df.loc[n - 1, "Anomaly Flag"] = "Yes (Negative Difference)"
```
